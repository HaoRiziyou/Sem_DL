```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# the origin file need to be preprocessed
path = './save/experiment-log.txt'
```

```python
reward = []
dis = []
epoch = []
train_loss = []
dis_loss = []
acc = []

with open(path) as f:
    for num, line in enumerate(f, 1):
        if "reward-train" in line:
            reward.append(line)
        if "discriminator" in line:
            dis.append(line)
            
for line in reward:
    epoch.append(line.split()[2])
    train_loss.append(line.split()[5])


for line in dis:
    dis_loss.append(line.split()[2])
    acc.append(line.split()[4])


# from the tenth element eof the dis loss and accuracy
df = pd.DataFrame({'epoch': epoch, 'train_loss':train_loss, 'dis_loss':dis_loss[5:], 'acc':acc[5:]})

target = './save/target.txt'

df.to_csv(target, sep=' ', index=False, header=False)
#save the file to target
```

```python
#open the target txt
target_="./save/target.txt"

with open(target_) as f:
    lines = f.readlines()
    x = [line.split()[0] for line in lines]
    y = [line.split()[1] for line in lines]
    

#with open("target.txt") as f:
with open(target_) as f:
    lines = f.readlines()
    a = [line.split()[0] for line in lines]
    b = [line.split()[2] for line in lines]

with open(target_) as f:
    lines = f.readlines()
    c = [line.split()[0] for line in lines]
    d = [line.split()[3] for line in lines]    

fig = plt.figure()
fig2 = plt.figure()
fig3 = plt.figure()

ax1 = fig.add_subplot()
ax2 = fig2.add_subplot()
ax3 = fig3.add_subplot()
ax1.set_title("Traing Curve")    
ax1.set_xlabel('epoch')
ax1.set_ylabel('train loss')

ax2.set_title("Discriminator Curve")    
ax2.set_xlabel('epoch')
ax2.set_ylabel('discriminator loss')

ax3.set_title("Accuracy")    
ax3.set_xlabel('epoch')
ax3.set_ylabel('Accuracy')

x_new_list =[float(i) for i in x]

y_new_list = [float(k) for k in y]
#print(x_new_list)
#print(y_new_list)
a_list = [float(i) for i in a]
b_list = [float(k) for k in b]

c_list = [float(i) for i in c]
d_list = [float(k) for k in d]

ax1.plot(x_new_list,y_new_list, label='train')
ax2.plot(a_list,b_list,label="discriminator")
ax3.plot(c_list,d_list,label="accuracy")

#leg = ax2.legend()
#leg = ax1.legend()
#leg = ax3.legend()
#x =[0,5,10,15,20]
#y= [10.707,15.227,6.1876,7.214,9.888]
#plt.plot(x,y)

plt.show()
```
